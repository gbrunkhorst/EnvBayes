# Plotting Logo

```python
import matplotlib.pyplot as plt
import piplite
await piplite.install(['seaborn'], keep_going=True)
import seaborn as sns
import scipy.stats as st
import numpy as np
cmap = sns.dark_palette("seagreen", reverse=True, as_cmap=True)
cmap
```

```python
sns.set_palette("Greens_r")
```

```python
fig, ax = plt.subplots(1, 1)

x = np.linspace(-5,5, 1000)
kappas = [2, 1/3]
for kappa in kappas:
    rv = st.laplace_asymmetric(kappa)
    ax.plot(x, rv.pdf(x), lw=3, label='frozen pdf')
#ax.text(-4,0.35,'EnvBayes')
ax.text(2,0.3,'Bayesian\nEnvironmental\nModeling')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)

x = np.linspace(-10,10, 1000)
kappas = [1.7, 1, 1/1.2]
offsets = [-2,0,1.3]
coeffs  = [.5,.3, .5]
heights = [.6, 1 ,.8]
for kappa, offset, coeff, height in zip(kappas, offsets, coeffs, heights):
    rv = st.laplace_asymmetric(kappa)
    ax.plot(x, height*rv.pdf(coeff*x+offset), lw=3, label='frozen pdf')
#ax.text(-4,0.35,'EnvBayes')
#ax.text(2,0.3,'Bayesian\nEnvironmental\nModeling')
plt.show()
```
